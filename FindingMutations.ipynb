```python
import numpy as np
import matplotlib.pyplot as plt
```

# AA Sequence: MLPTALKTLGKKCKKSPLVVRNIVPGKPGTTKPTSLVMTAAINGQALSLV

```python
with open("/Users/kevin/Downloads/sequence_1.txt") as f:
    lines = f.readlines()
    # Count holds number of mutations at each position (50 positions for gene 1)
    Count = {}
    [Count.setdefault(x, 0) for x in range(50)]
    
    non_mutations = ["*", " ", "."]
    
    for line in lines:
        # indices 19-69 hold the AA sequence
        sequence = line[19:69]
        for i in range(len(sequence)):
            if sequence[i] not in non_mutations:
                Count[i] += 1
                
    # holds number of positions with 0 mutations
    identical_positions = 0
    for key, value in Count.items():
        print("Position " + str(key+1) + " is mutated in " + str(value) + " out of " + str(len(lines)) + " sequences")
        
        if value == 0:
            identical_positions += 1
            
    print("\n" + str(identical_positions) + " / 50 Identity")   
```

```python
x_labels = list("MLPTALKTLGKKCKKSPLVVRNIVPGKPGTTKPTSLVMTAAINGQALSLV")
x = Count.keys()
y = Count.values()

plt.figure(figsize=[20,10])

plt.bar(range(len(y)), y, align='center')
plt.xticks(range(len(x)), x_labels, size='small', fontsize = 15)
plt.yticks(fontsize = 15)

plt.title("Frequency of mutations across all available SARS-Cov-2 genomes", fontsize=25)
plt.xlabel('Amino Acid in Query by Position',fontsize=25)
plt.ylabel('Mutation Frequency',fontsize=25)
plt.savefig('gene1.png')
```
